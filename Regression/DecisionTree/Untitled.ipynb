```python
import pandas as pd
```

```python
data1=pd.read_csv("50_Startups.csv")
data1
```

```python
data=pd.get_dummies(data1, drop_first='TRUE')
```

```python
data
```

```python
data.columns
```

```python
x=data[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
y=data[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson',splitter='best')
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_pred,y_test)
r2_score
```

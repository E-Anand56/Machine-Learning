```python
import pandas as pd 
```

```python
data1=pd.read_csv('Mall_Customers.csv')
```

```python
x=data1.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt

ap = AffinityPropagation(random_state=42)
ap.fit(x)

# 2 Get cluster centers and labels
cluster_centers = ap.cluster_centers_indices_
labels = ap.labels_
print(labels)

print("Cluster centers indices:", cluster_centers)
print("Number of clusters:", len(cluster_centers))

# 3 Optional: visualize if 2D data
if x.shape[1] == 2:
    plt.figure(figsize=(8,6))
    for cluster_id in set(labels):
        plt.scatter(x[labels == cluster_id, 0], x[labels == cluster_id, 1], label=f'Cluster {cluster_id}')
    plt.scatter(x[cluster_centers, 0], x[cluster_centers, 1], s=200, c='red', marker='x', label='Exemplars')
    plt.legend()
    plt.show()
```

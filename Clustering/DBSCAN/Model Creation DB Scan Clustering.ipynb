```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Step 2: Load the dataset
data1 = pd.read_csv('Mall_Customers.csv')

# Step 3: Select features (Annual Income and Spending Score)
x = data1.iloc[:, [3, 4]].values  # Columns 3 and 4

# Step 4: Feature scaling (important for DBSCAN)
x = StandardScaler().fit_transform(x)

# Step 5: Create and fit DBSCAN model
DB = DBSCAN(eps=0.3, min_samples=5)  # try tuning eps and min_samples
DB.fit(x)

# Step 6: Get cluster labels
labels = DB.labels_
print("Cluster labels:", labels)

# Step 7: Visualize clusters
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering (Mall Customers)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.show()

# Step 8 (optional): Count clusters
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Number of clusters found:", n_clusters)
```
